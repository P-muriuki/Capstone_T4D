```python
df.rename(columns={'B.O_rate_unknown_rider':'BO_rate_unknown_rider', 'B.O_rate_incompetent_rider': 'BO_rate_incompetent_rider', 'B.O_rate_accessibility_of_rider':'BO_rate_accessibility_of_rider'}, inplace = True)
```

```python
lencoder = LabelEncoder()
rencoder = LabelEncoder()
dencoder = LabelEncoder()
pencoder = LabelEncoder()
```

```python
df.Willing_to_pay  = lencoder.fit_transform(df.Willing_to_pay)
df.Opinion_Tracking_goods	 = rencoder.fit_transform(df.Opinion_Tracking_goods)
df.Whether_smallbusinessowner_ornot = dencoder.fit_transform(df.Whether_smallbusinessowner_ornot)
df.Will_to_pay_to_access_riders = pencoder.fit_transform(df.Will_to_pay_to_access_riders)
```

```python
new = df["Type_of_goods_delivered"].str.split(",", expand = True)
```

```python
df['Type_of_goods_Option1'] = new[0]
df['Type_of_goods_Option2'] = new[1]
df['Type_of_goods_Option3'] = new[2]

#The 3 columns are at the back.
```

```python
toncoder = LabelEncoder()
rotcoder = LabelEncoder()
wowcoder = LabelEncoder()

df.Type_of_goods_Option1  = toncoder.fit_transform(df.Type_of_goods_Option1)
df.Type_of_goods_Option2 = rotcoder.fit_transform(df.Type_of_goods_Option2)
df.Type_of_goods_Option3 = wowcoder.fit_transform(df.Type_of_goods_Option3)
```

```python
df = pd.get_dummies(df, columns = ['Importance_of_fast_delivery'])
```
